#loading and filtering data#

```python
import pandas as pd
```

```python
filename ='/content/RailsToTrails_National Count Data_week - Total Count data (31 counters) (2).csv'
bike_df = pd.read_csv(filename)
bike_df
```

-Selecting just the data after July 31, 2020 and saving it as a variable such as df_filtered

bike_df["Week of"] = pd.to_datetime(bike_df["Week of"]) to convert "week_of" to a datetime data type.  Then you can use > to filter the data using that column.

```python
bike_df["Week of"] = pd.to_datetime(bike_df["Week of"])
df_filtered = bike_df.loc[bike_df['Week of'] >'2020-07-31']
df_filtered
```

- Find how many entries occurred after July 31, 2020 Hint: use len(df_filtered) to get the length of the dataframe 

```python
len(df_filtered)
```

- Select just data where the 'Change 2019 -2020' column is greater than 50% (greater than .5).

```python
bike_df.loc[bike_df['Change 2019-2020']>0.5]
```

Put these together! When filtering you can use "&" to combine two filters. Determine how many weeks in the second half of the year (after July 31, 2020) also had a change greater than 50%? Hint: use len().

```python
filter= bike_df['Change 2019-2020']>0.5
bike_df.loc[df_filtered & filter]
```
